```python
#Bibliotecas utilizadas para a analise financeira de ativos
import yfinance as yf
import pandas as pd
import numpy as np
import investpy
import requests
from statsmodels.tsa.stattools import coint
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
#Download de dados historicos de fechamento de todos os ativos

ibov = yf.download("^BVSP", start="2015-01-01", end="2025-10-05")['Close']

sp500 = yf.download("^GSPC", start="2015-01-01", end="2025-10-05")['Close']

vix = yf.download("^VIX", start="2015-01-01", end="2025-10-05")['Close']

usdb = yf.download("USDBRL=X", start="2015-01-01", end="2025-10-05")['Close']

ouro = yf.download("GC=F", start="2015-01-01", end="2025-10-05")['Close']

url = 'https://api.bcb.gov.br/dados/serie/bcdata.sgs.433/dados?formato=json'
res = requests.get(url).json()
ipca = pd.DataFrame(res)
ipca['data'] = pd.to_datetime(ipca['data'], dayfirst=True)
ipca['valor'] = pd.to_numeric(ipca['valor'])
ipca.set_index('data', inplace=True)
ipca = ipca[ipca.index >= '2015-01-01']

print(ipca.head())
```

```python
#Retorno de logaritmo dos ativos em 1 dia
ibov['ret_log'] = np.log(ibov['^BVSP'] / ibov['^BVSP'].shift(1))
sp500['ret_log'] = np.log(sp500['^GSPC'] / sp500['^GSPC'].shift(1))
vix['ret_log'] = np.log(vix['^VIX'] / vix['^VIX'].shift(1))
usdb['ret_log'] = np.log(usdb['USDBRL=X'] / usdb['USDBRL=X'].shift(1))
ouro['ret_log'] = np.log(ouro['GC=F'] / ouro['GC=F'].shift(1))
ipca['ret_log'] = np.log(ipca['valor'] / ipca['valor'].shift(1))

#Retorrno de logaritmo dos ativos em 1 mês
ibov['ret_mensal'] = ibov['ret_log'].rolling(window=28, min_periods=1).sum()
sp500['ret_mensal'] = sp500['ret_log'].rolling(window=28, min_periods=1).sum()
vix['ret_mensal'] = vix['ret_log'].rolling(window=28, min_periods=1).sum()
usdb['ret_mensal'] = usdb['ret_log'].rolling(window=28, min_periods=1).sum()
ouro['ret_mensal'] = ouro['ret_log'].rolling(window=28, min_periods=1).sum()
ipca['ret_mensal'] = ipca['ret_log'].rolling(window=28, min_periods=1).sum()

#Retorno de logaritmo dos ativos em 1 ano
ibov['ret_anual'] = ibov['ret_log'].rolling(window=365, min_periods=1).sum()
sp500['ret_anual'] = sp500['ret_log'].rolling(window=365, min_periods=1).sum()
vix['ret_anual'] = vix['ret_log'].rolling(window=365, min_periods=1).sum()
usdb['ret_anual'] = usdb['ret_log'].rolling(window=365, min_periods=1).sum()
ouro['ret_anual'] = ouro['ret_log'].rolling(window=365, min_periods=1).sum()
ipca['ret_anual'] = ipca['ret_log'].rolling(window=365, min_periods=1).sum()
   
```

```python
#Volatilidade de ativos 1 dia 
ibov['Volatilidade_diaria']=ibov['ret_log']**2
sp500['Volatilidade_diaria']=sp500['ret_log']**2
vix['Volatilidade_diaria']=vix['ret_log']**2
usdb['Volatilidade_diaria']=usdb['ret_log']**2
ouro['Volatilidade_diaria']=ouro['ret_log']**2
ipca['Volatilidade_diaria']=ipca['ret_log']**2

#Volatilidade de ativos 1 mês
ibov['Volatilidade_21d'] = ibov['ret_log'].rolling(window=21).std()
sp500['Volatilidade_21d'] = sp500['ret_log'].rolling(window=21).std()
vix['Volatilidade_21d'] = vix['ret_log'].rolling(window=21).std()
usdb['Volatilidade_21d'] = usdb['ret_log'].rolling(window=21).std()
ouro['Volatilidade_21d'] = ouro['ret_log'].rolling(window=21).std()
ipca['Volatilidade_21d'] = ipca['ret_log'].rolling(window=21).std()

#Volatilidade de ativos 1 ano 

ibov['Volatilidade_21d'] = ibov['ret_log'].rolling(window=365).std()
sp500['Volatilidade_21d'] = sp500['ret_log'].rolling(window=365).std()
vix['Volatilidade_21d'] = vix['ret_log'].rolling(window=365).std()
usdb['Volatilidade_21d'] = usdb['ret_log'].rolling(window=365).std()
ouro['Volatilidade_21d'] = ouro['ret_log'].rolling(window=365).std()
ipca['Volatilidade_21d'] = ipca['ret_log'].rolling(window=365).std()

print(ipca['Volatilidade_diaria'].head())
```

```python
#correlação entre os ativos 
data = pd.concat([ibov['ret_log'], sp500['ret_log'], vix['ret_log'], usdb['ret_log'], ouro['ret_log'], ipca['ret_log']], axis=1)
data.columns = ['IBOV', 'SP500', 'VIX', 'USDBRL', 'OURO', 'IPCA']
correlation_matrix = data.corr()    
```
